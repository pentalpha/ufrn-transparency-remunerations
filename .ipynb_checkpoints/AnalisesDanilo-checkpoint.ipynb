```python
import pandas as pd
from bokeh.charts import Scatter, BoxPlot, output_file, show, output_notebook
from bokeh.layouts import row
genderDF = pd.read_csv("csv/serversTreatedWithGender.csv")
dfWithoutGenderX = genderDF[genderDF['gender'] != 'X']
```

# Analysis per level/class/department(or organization)
# Salary women vs men per level

```python
#p1 = Scatter(genderDF, x = 'money', y = 'level', color = 'gender', title = 'Wage relation Men(M) x Women(F) per level')

#p2 = Scatter(genderDF.sort_index(by = 'class', ascending=False), x = 'money', y = 'class', color = 'gender', title = 'Wage relation Men(M) x Women(F) per class')

p1 = Scatter(dfWithoutGenderX, x = 'money', y = 'level', color = 'gender', marker = 'gender', title = 'Wage relation Men(M) x Women(F) per level')

p2 = Scatter(dfWithoutGenderX.sort_index(by = ['class', 'gender'], ascending=False), x = 'money', y = 'class', color = 'gender', marker = 'gender', title = 'Wage relation Men(M) x Women(F) per class')

layout = row(p1, p2, sizing_mode='scale_width')

#output_file('scatter.html')
output_notebook()

show(layout)
```

# Wage relation Men(M) x Women(F) by occupation (>30000) and Wage relation Men(M) x Women(F) organization/department (>30000)

```python
dftest = dfWithoutGenderX[dfWithoutGenderX['money'] > 30000]

#p1 = Scatter(genderDF, x = 'money', y = 'level', color = 'gender', title = 'Wage relation Men(M) x Women(F) per level')

#p2 = Scatter(genderDF.sort_index(by = 'class', ascending=False), x = 'money', y = 'class', color = 'gender', title = 'Wage relation Men(M) x Women(F) per class')

p1 = Scatter(dftest, x = 'money', y = 'post', color = 'gender', marker = 'gender', title = 'Wage relation Men(M) x Women(F) occupation (>30000)')

p2 = Scatter(dftest.sort_index(by = ['class', 'gender'], ascending=False), x = 'money', y = 'org', color = 'gender', marker = 'gender', title = 'Wage relation Men(M) x Women(F) org (>30000)')

layout = row(p1, p2, sizing_mode='scale_width')

#output_file('scatter2.html')
output_notebook()
show(layout)
```

# IMD 

```python
dfIMD = dfWithoutGenderX[dfWithoutGenderX['org'] == 'INSTITUTO METROPOLE DIGITAL']

print("Number of functionaries in IMD: " + str(len(dfIMD)))
print("Number of women in IMD: " + str(len(dfIMD[dfIMD['gender'] == 'F'])))
print("Number of men in IMD: " + str(len(dfIMD[dfIMD['gender'] == 'M'])))

p = BoxPlot(dfIMD, values = 'money', label = 'gender', color = 'gender', xlabel = 'Gender', ylabel = 'money')

output_notebook()

show(p)
```

# Conclusions
Here the analysis was focused on levels and departments. The differences between women and men are not too distant, but there is a little advantage of men in relation to women.

By level, always men are earning more than women. This pattern is similar (but not always) when by class.

When comparing department / organization there is almost equallity between genres and finally, then analysing just IMD we see that there is a large difference between number of men and women. The men salaries are greater than salaries women, but the difference is not so big.
