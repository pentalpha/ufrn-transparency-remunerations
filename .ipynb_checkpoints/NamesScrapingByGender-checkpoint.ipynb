```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from threading import Thread
base = 'csv/'
```

```python
# url : url of a page
# returns: content of page requested (html)
def requestHTMLPage(url):
    r = requests.get(url)
    while r.status_code != 200:
        time.sleep(10)
        r = requests.get(url)
        print("Loop em potencial")
    return r.content
```

```python
# urls where there are female names and male names
baseUrlFemale = "https://www.dicionariodenomesproprios.com.br/nomes-femininos/"
maxNumberFemale = 151 # total of pages with female names
baseUrlMale = "https://www.dicionariodenomesproprios.com.br/nomes-masculinos/"
maxNumberMale = 193 # total of pages with male names
```

```python
# maxNumber : total of pages will be accessed to get names
# urlBase : where the names were found
# returns : list of string with names
def getNames(maxNumber, urlBase):
    namesByGenre = []
    for i in range(1, maxNumber + 1):
        c = requestHTMLPage(urlBase + str(i))
        soup = BeautifulSoup(c, "html.parser")
        names = soup.find_all('a', class_="lista-nome")
        for n in names:
            namesByGenre.append(n.getText().strip())
        print("Current page " + str(i))
    return namesByGenre
```

```python
# saves in fineName the given names
def saveCSV(fileName, names):
    file = open(base + fileName, 'w')
    for n in names:
        file.write(n + '\n')
    file.close()
```

```python
# get female names
female = getNames(maxNumberFemale, baseUrlFemale)
female = sorted(female)
```

```python
saveCSV("femaleNames.csv", female)
```

```python
# get male names
male = getNames(maxNumberMale, baseUrlMale)
male = sorted(male)
```

```python
saveCSV("maleNames.csv", male)
```
