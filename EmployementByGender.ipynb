# Employement by Gender at UFRN

```python
import pandas as pd
import numpy as np
from bokeh.charts import Donut, show, output_notebook

df = pd.read_csv("csv/serversTreatedWithGender.csv")
df.head()
```

```python
genderDF = df[df.gender != 'X']
genderDF.head()
```

```python
genderDF['money'] = pd.to_numeric(genderDF['money'], errors='coerce').fillna(0)
#genderDF.head()
#print(str(genderDF['money'].mean()))
donut = Donut(genderDF, label=['org','gender'])
output_notebook()
show(donut)
```
